```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import plot, savefig
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python

X, y = make_regression(n_samples=50, n_features=1, noise=10, random_state=42)
```

```python
X
```

```python
model = LinearRegression().fit(X, y)
x_line = np.linspace(X[:, 0].min(), X[:, 0].max(), 200)
y_line = model.predict(x_line.reshape(-1, 1))
y_pred = model.predict(X)
```

```python
fig, ax = plt.subplots(1, 1)

fig.set_figheight(7)
fig.set_figwidth(15)

ax.scatter(X[:, 0], y, marker='.')
ax.scatter(x_line, y_line, marker='.')

for xi, yi, yhat in zip(X[:, 0], y, y_pred):
    plt.plot([xi, xi], [yi, yhat], linestyle="--")

ax.legend(["Data", "Fitted Function", "Error Visualization"])

mse = np.mean((y - y_pred) ** 2)
mae = np.mean(abs(y - y_pred))
# ax.set_title(f"MSE = {mse:.4f}\nMAE = {mae:.4f}")

savefig('MSE_MAE_Plot.png', transparent=True)
```
